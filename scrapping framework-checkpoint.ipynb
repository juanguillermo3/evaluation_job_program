```python

#
def cedulas_ya_revisadas(parametros_de_persistencia):
    #
    import os
    _,_,files=list(os.walk(
        parametros_de_persistencia["directorio_tabla"]
    ))[0]
    os.chdir(parametros_de_persistencia["directorio_tabla"])
    #
    import re
    files=[f for f in files if re.search(parametros_de_persistencia["nombre_tabla"], f )]
    #
    cedulas=[]
    for f in files:
        import pandas as pd
        try:
            df=pd.read_csv(f)
            df.columns=[c.lower() for c in df.columns]
            for c in df["cedula"]:
                cedulas.append(str(c))
        except:
            pass
    #
    return(list(set(cedulas)))
#
#cedulas_ya_revisadas(
#    parametros_de_persistencia={
#        "nombre_tabla":"procesos laborales",
#        "directorio_tabla":PROCESOS_HOME
#    }
#)
#
cedulas_ya_revisadas(
    parametros_de_persistencia={
        "nombre_tabla":"scraper delivery",
        "directorio_tabla":SERVICIOS_HOME
    }
)
```
